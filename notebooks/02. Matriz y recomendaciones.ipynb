## Similitud de coseno

Se calcula la matriz completa de similitud de coseno que indica que tan similares son dos vectores. En este caso si dos pel√≠culas tienen vectores similares se espera que sean pel√≠culas similares. Se usa la formula del cuaderno 01, pero en este caso se aplica a la matriz completa para obtener un valor para cada indice y columna que corresponde a la similitud entre la pel√≠cula indice y al pel√≠cula columna. Para el c√°lculo nos apoyamos de la librer√≠a 

```python
df_similitud = pd.DataFrame(1 - cdist(df_movies, df_movies, metric='cosine'))
df_similitud.columns, df_similitud.index = list(df_movies.index), list(df_movies.index)
```

En la matriz cada valor i,j indica que tan parecido es el vector de la pel√≠cula i con el de la pel√≠cula j. Si este valor se acerca a 1 son muy similares. Claramente una pelicula comparada con s√≠ misma tendr√° un valor de 1. Las recomendaciones corresponden a las pel√≠culas con el valor de similitud m√°s alto.

```python
df_similitud.to_csv('../models/similarity.csv', index = True)
```

## Recomendaci√≥n

A continuaci√≥n se muestran algunos ejemplos de recomendaciones de pel√≠culas.

```python
def recomendar(pelicula, cantidad):
    print(f'''
    Usted vio la pel√≠cula {pelicula}.

    Sus recomendaciones son las siguientes:
    ''')

    for index, movie in enumerate(df_similitud[pelicula].sort_values(ascending=False)[1:cantidad+1].index):
        print(str(index+1) + '. ' + movie)
```

```python
PELICULA_VISTA = 'The Avengers'
CANTIDAD_DE_RECOMENDACIONES = 10
recomendar(PELICULA_VISTA, CANTIDAD_DE_RECOMENDACIONES)
```
