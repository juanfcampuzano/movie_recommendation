## Similitud de coseno

Se calcula la matriz completa de similitud de coseno que indica que tan similares son dos vectores. En este caso si dos películas tienen vectores similares se espera que sean películas similares. Se usa la formula del cuaderno 01, pero en este caso se aplica a la matriz completa para obtener un valor para cada indice y columna que corresponde a la similitud entre la película indice y al película columna. Para el cálculo nos apoyamos de la librería 

```python
df_similitud = pd.DataFrame(1 - cdist(df_movies, df_movies, metric='cosine'))
df_similitud.columns, df_similitud.index = list(df_movies.index), list(df_movies.index)
```

En la matriz cada valor i,j indica que tan parecido es el vector de la película i con el de la película j. Si este valor se acerca a 1 son muy similares. Claramente una pelicula comparada con sí misma tendrá un valor de 1. Las recomendaciones corresponden a las películas con el valor de similitud más alto.

```python
df_similitud.to_csv('../models/similarity.csv', index = True)
```

## Recomendación

A continuación se muestran algunos ejemplos de recomendaciones de películas.

```python
def recomendar(pelicula, cantidad):
    print(f'''
    Usted vio la película {pelicula}.

    Sus recomendaciones son las siguientes:
    ''')

    for index, movie in enumerate(df_similitud[pelicula].sort_values(ascending=False)[1:cantidad+1].index):
        print(str(index+1) + '. ' + movie)
```

```python
PELICULA_VISTA = 'The Avengers'
CANTIDAD_DE_RECOMENDACIONES = 10
recomendar(PELICULA_VISTA, CANTIDAD_DE_RECOMENDACIONES)
```
